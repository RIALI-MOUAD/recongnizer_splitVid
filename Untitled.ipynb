```python
def find_speaking(audio_clip, window_size=0.1, volume_threshold=0.4, ease_in=0.25):
    # First, iterate over audio to find all silent windows.
    num_windows = math.floor(audio_clip.end/window_size)
    window_is_silent = []
    for i in range(num_windows):
        s = audio_clip.subclip(i * window_size, (i + 1) * window_size)
        v = s.max_volume()
        window_is_silent.append(v < volume_threshold)

    # Find speaking intervals.
    speaking_start = 0
    speaking_end = 0
    speaking_intervals = []
    for i in range(1, len(window_is_silent)):
        e1 = window_is_silent[i - 1]
        e2 = window_is_silent[i]
        # silence -> speaking
        if e1 and not e2:
            speaking_start = i * window_size
        # speaking -> silence, now have a speaking interval
        if not e1 and e2:
            speaking_end = i * window_size
            new_speaking_interval = [speaking_start - ease_in, speaking_end + ease_in]
            # With tiny windows, this can sometimes overlap the previous window, so merge.
            need_to_merge = len(speaking_intervals) > 0 and speaking_intervals[-1][1] > new_speaking_interval[0]
            if need_to_merge:
                merged_interval = [speaking_intervals[-1][0], new_speaking_interval[1]]
                speaking_intervals[-1] = merged_interval
            else:
                speaking_intervals.append(new_speaking_interval)

    return speaking_intervals
```

```python
import json

with open('sample.json') as json_file:
    data = json.load(json_file)

chunks = glob.glob('frames/zAjJYkUnTEs/chunk*')


def get_stri(i,chunks=chunks, data=data):
    indx = chunks[i].split('chunk')[-1]
    data[indx]
    elmnt = data[indx]['N.B']
    try:
        audio, text, video = elmnt['Audio'], elmnt['Text'], elmnt['Video']
        return f'Audio :{audio} | Text :{text} | Video :{video}'
    except:
        return 'Not Detected :)'
```

```python
#Write on frames
def write_on_frame(path,
                   out,
                   text,
                   font = cv2.FONT_HERSHEY_SIMPLEX,
                   fontScale = 1,
                   fontColor = (245, 245, 220),
                   thickness = 2,
                   lineType = 2):
    img = cv2.imread(path)
    shape = img.shape
    # Write some Text
    bottomLeftCornerOfText = (shape[1]//20,shape[0]//10)
    cv2.putText(img,text,
                bottomLeftCornerOfText,
                font,
                fontScale,
                fontColor,
                thickness,
                lineType)

    #Display the image
    #cv2.imshow("img",img)
    #Save image
    cv2.imwrite(out, img)
    cv2.waitKey(0)
```

```python
chunks = glob.glob('frames/zAjJYkUnTEs/chunk*')
chunks.sort(key=lambda x: int(''.join(filter(str.isdigit, x))))
def write_on_frames(chunk,portion = 0.5, tresh = 45):
    frames = glob.glob(f'frames/zAjJYkUnTEs/chunk{chunk}/frame*')
    frames.sort(key=lambda x: int(''.join(filter(str.isdigit, x))))
    n = min(len(frames),tresh)
    text = get_stri(chunk)
    for frame in frames[len(frames)-n:]:
        write_on_frame(frame,frame,text)
    return frames
        
frames_glob = []
for i in range(len(chunks)):
    frames_loc = write_on_frames(i)
    frames_glob+=frames_loc
```

```python
def write_video(file_path, frames, fps):
    """
    Writes frames to an mp4 video file
    :param file_path: Path to output video, must end with .mp4
    :param frames: List of PIL.Image objects
    :param fps: Desired frame rate
    """
    img0 = cv2.imread(frames[0])
    w, h = img0.shape[1],img0.shape[0]
    print(w,h)
    fourcc = cv2.VideoWriter_fourcc('m', 'p', '4', 'v')
    writer = cv2.VideoWriter(file_path, fourcc, fps, (w, h))

    for frame,i in zip(frames,tqdm(range(len(frames)))):
        frame = cv2.imread(frame)
        writer.write(frame)

    writer.release() 
```
